```
#1.8
#Caja
ancho <- 9
alto <- 10

for (i in 1:alto){
  if(i == 1 || i == alto){cat(strrep("*",ancho),"\n")}
  else{cat(paste0("*",strrep(" ",ancho-2), "*\n"))}
}
cat("\n")

#Ovalo
ancho <-7
alto <-9

#Deja a proposito un espacio entre los asteriscos arriba/abajo y los de enmedio
for (i in 1:alto){
  if(i == 1 || i == alto){cat(strrep(" ",2),strrep("*",ancho-4),"\n")}          #se puede poner cat(paste0(code))
  else if(i == 2 || i == alto-1){cat(paste0(" *",strrep(" ",ancho-2), "* \n"))} #se puede cambiar a ancho-4
  else {cat(paste0("*",strrep(" ",ancho), "*\n"))}                              #se puede cambiar a ancho-
}
cat("\n")

#Flecha
ancho <- 5
altura <- 9

alturahead <- 0                        
blank <- (ancho-1)
while (blank >=0){
  cat(paste0(strrep(" ",blank/2), strrep("*",ancho-blank),"\n"))
  alturahead = alturahead+1                        
  blank = blank -2
}
while (alturahead <= altura){
  cat(paste0(strrep(" ",(ancho-1)/2),"*\n"))
  alturahead = alturahead+1
}
cat("\n")

#Diamante
arista <- 9

blank <- (arista-1)/2
between <- 1
half = ceiling(arista/2)

for (x in 1:arista){
  if(x==1 || x == arista){
    cat(paste0(strrep(" ",blank),"*\n"))
    blank <- blank-1
    }
  else {
      if(x<half){
        cat(paste0(strrep(" ",blank),"*",strrep(" ",between),"*\n"))
        blank <- blank-1
        between <- between+2
      }
      else{
        cat(paste0(strrep(" ",blank),"*",strrep(" ",between),"*\n"))
        blank <- blank+1
        between <- between-2
      }
    }
  }
```

```
#1.10
cat("Ingrese la altura: ")
{  
  altura <- as.integer(readline(prompt = ""))
  ancho <- altura*2 -1
  
  blank <- (ancho-1)
  while (blank >=0){
    cat(paste0(strrep(" ",blank/2), strrep("*",ancho-blank),"\n"))
    blank = blank -2
  }
}
```
