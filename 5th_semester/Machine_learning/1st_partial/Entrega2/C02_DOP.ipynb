```
valores = str()
for i in range(1,11):
  valores += str(i) + ' '
print(valores)

x = np.linspace(-math.pi,math.pi,50)
```

```
y = x
def function1(x,y):
  return math.cos(y)/(1 + x**2)

z = np.zeros((len(x),len(y)))
for i in range(len(y)):
  for k in range(len(x)):
    z[i][k] = function1(x[k],y[i])

plt.contour(x, y, z)
plt.show()
plt.contour(x, y, z,45)
plt.show()


def function2(x,y):
  return (function1(x,y) - np.transpose(function1(x,y)) / 2)

ztras = np.zeros((len(x),len(y)))
for i in range(len(y)):
  for k in range(len(x)):
    ztras[i][k] = function2(x[k],y[i])

plt.contour(x, y, ztras,15)
plt.show()
```

```
heat_map = sb.heatmap(ztras)
plt.show()
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x,y,ztras,100)
```
